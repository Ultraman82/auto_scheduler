```python
from datetime import datetime, time
import os
import pandas as pd
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)
```

```python
machine_path = "./data_storage/mc_machine.xlsx"
machine_list = pd.read_excel(machine_path)
machine_list['variant'] = ''
hsr_variant = machine_list.columns[2:10]
else_variant = machine_list.columns[10:-1]
for row in machine_list.itertuples():    
    type_list = row.type.split(", ")
    variants = ""
    for type_ in type_list:
        if "HSR" in type_:
            for i, v in enumerate(row[3:11]):
                if v == 0:
                    variants += type_ + hsr_variant[i] + " "
        else:
            for i, v in enumerate(row[11:-1]):
                if v == 0:
                    variants += type_ + else_variant[i] + " "
    machine_list.loc[row[0], 'variant'] = variants
```
